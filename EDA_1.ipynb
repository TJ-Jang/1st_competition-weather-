```python
gs25.columns
```

```python
gs25 = gs25.rename(columns = {'korea_cvs.pvn_nm' : 'pvn_nm',
                              'korea_cvs.sale_dt' : 'tm',
                              'korea_cvs.gen_cd' : 'gen_cd',
                              'korea_cvs.age_cd' : 'age_cd',
                              'korea_cvs.category' : 'category',
                              'korea_cvs.adj_qty' : 'qty'})
```

```python
gs25.head()
```

```python
gs25 = gs25.drop('korea_cvs.bor_nm', axis = 1)
```

```python
gs25 = pd.DataFrame(gs25, columns = ('tm', 'pvn_nm', 'gen_cd', 'age_cd', 'category' ,'qty'))
```

```python
gs25.head()
```

  * gs25를 weather과 merge 시키기 위해서, Mechine Learning모델을 넣기 위해서는 데이터를 맞게 조금 변형할 필요가 있다. 지금은 그 사전처리만 해 둘 예정이다.

```python
# gen_cd 자료 중 F = 0, M = 1로 변환
gs25['gen_cd'] = gs25['gen_cd'].map({'F':0, 'M':1})
```

```python
# age_cd 자료는 분류하기 쉽도록 되어있어 편하게 바꿀 수 있다.
gs25['age_cd'].unique()
```

```python
def age_category(x):
    if x == '00~19':
        return 0
    elif x == '20~39':
        return 1
    elif x == '40~59':
        return 2
    else:
        return 3

gs25['age_cd'] = gs25['age_cd'].apply(age_category)
```

```python
# born_nm 데이터가 사라짐에 따라 qty의 데이터에도 변화가 필요하다. 평균데이터를 넣어도 좋으나 
# sum을 넣어 활용하는 것이 더 좋을 것으로 보인다.
# 이를 더욱 편하게 하기 위하여 사전에 category를 숫자화시키겠다.
gs25['category'].unique()
```

```python
gs25['category'] = gs25['category'].map({'라면':100,
                                         '과자':101,
                                         '마스크':102,
                                         '맥주':103,
                                         '생리대':104,
                                         '생수':105,
                                         '숙취해소제':106,
                                         '스타킹':107,
                                         '아이스크림':108,
                                         '탄산음료':109,
                                         '면도기':110,
                                         '우산':111})
```

```python
# 사전준비는 완료되었다. 이제 gs25의 qty_sum 데이터를 만들도록 하겠다.
for t in tm:
    for a in gs25['pvn_nm'].unique():
        for b in range(2): # gen_cd
            for c in range(4): # age_cd
                for d in gs25['category'].unique():
                    gs25.loc[(gs25.tm==t)&(gs25.pvn_nm == a)&(gs25.gen_cd == b)&(gs25.age_cd == c)&(gs25.category == d),'qty'] = gs25.loc[(gs25.tm==t)&(gs25.pvn_nm == a)&(gs25.gen_cd == b)&(gs25.age_cd == c)&(gs25.category == d),'qty'].sum()
    print(t, 'complete')
```

```python
gs25 = gs25.drop_duplicates()
```
